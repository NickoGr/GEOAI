# GeoAI audit reproducibility notebook

This notebook consolidates the analysis artifacts used for the final reliability section and Table 5 checks. It is intended as a portable `.ipynb` file for versioning in Git and for updating the repository-linked reproducibility materials.

## Included files

The notebook reads the CSV outputs currently available in `output/` and shows the key values used in the manuscript.

```python
from pathlib import Path
import pandas as pd

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 200)

BASE = Path('output')
FILES = [
    'table5_scores.csv',
    'icc_pre_post_humans.csv',
    'icc_configurations.csv',
    'icc_pre_post_by_layer.csv',
    'icc_by_layer_all6.csv',
    'pre_post_changed_items.csv',
]

existing = [f for f in FILES if (BASE / f).exists()]
missing = [f for f in FILES if not (BASE / f).exists()]
existing, missing
```

## File schemas

Headers detected while packaging the notebook:

- **table5_scores.csv**: Case, Layer, E1, E2, E3, Claude, GPT_5_4, Nemotron
- **icc_pre_post_humans.csv**: configuration, n_items, k_raters, ICC2_1, ICC2_k, F, df1, df2, p, MSR, MSC, MSE
- **icc_configurations.csv**: configuration, n_items, k_raters, ICC2_1, ICC2_k, F, df1, df2, p, MSR, MSC, MSE
- **icc_pre_post_by_layer.csv**: Layer, Pre_ICC2k, Pre_p, Post_ICC2k, Post_p
- **icc_by_layer_all6.csv**: layer, n_items, k_raters, ICC2_1, ICC2_k, F, df1, df2, p
- **pre_post_changed_items.csv**: Case, Layer, n_changed, pre_range, post_range, pre_scores, post_scores

```python
dfs = {f: pd.read_csv(BASE / f) for f in existing}
for name, df in dfs.items():
    print(f'\n=== {name} ===')
    display(df.head(20))
```

## Key manuscript checks

This cell extracts the headline reliability values used in the revised text.

```python
headline = {}

if 'icc_pre_post_humans.csv' in dfs:
    headline['pre_post_humans'] = dfs['icc_pre_post_humans.csv']

if 'icc_configurations.csv' in dfs:
    headline['configurations'] = dfs['icc_configurations.csv']

headline
```

```python
for key, df in headline.items():
    print(f'\n### {key}')
    display(df)
```

## Optional consistency check

This cell looks for the GPT-5.4 and full six-rater rows so you can verify the distinction directly before updating the repository materials.

```python
if 'icc_configurations.csv' in dfs:
    cfg = dfs['icc_configurations.csv'].copy()
    display(cfg[cfg.astype(str).apply(lambda s: s.str.contains('GPT|six|6|all', case=False, regex=True)).any(axis=1)])
else:
    print('icc_configurations.csv not found')
```
